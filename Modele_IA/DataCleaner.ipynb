### Import librairies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import metrics
from matplotlib import rcParams
rcParams['figure.figsize'] = 11.7,8.27
```

### Import dataset

```python
df = pd.read_json(r"data_harvest_full.json", lines=True)
df.head()
```

```python
df_clean=df.drop(['status'],axis=1)
```

```python
df_clean.info()
```

```python
for i in range(0, len(df_clean)): 
    df_clean.iloc[i].fuel_type = df_clean.iloc[i].fuel_type[8:] 

for i in range(0, len(df_clean)): 
    df_clean.iloc[i].emission = df_clean.iloc[i].emission[17:-5] 

for i in range(0, len(df_clean)): 
    df_clean.iloc[i].transmission = df_clean.iloc[i].transmission[13:] 

for i in range(0, len(df_clean)): 
    df_clean.iloc[i].fuel_usage = df_clean.iloc[i].fuel_usage[19:-16] 

for i in range(0, len(df_clean)): 
    df_clean.iloc[i].url = df_clean.iloc[i].url[44:] 

for i in range(0, len(df_clean)): 
    df_clean.iloc[i].version = df_clean.iloc[i].version[8:] 

for i in range(0, len(df_clean)): 
    df_clean.iloc[i].year = df_clean.iloc[i].year[:] 

df_clean['door_nb'] = df_clean['door_nb'].str.replace(r'\D', '')
df_clean['seat_nb'] = df_clean['seat_nb'].str.replace(r'\D', '')
df_clean['location'] = df_clean['location'].str.replace(r'\D', '')
df_clean['price'] = df_clean['price'].str.replace(r'\D', '')
df_clean['mileage'] = df_clean['mileage'].str.replace(r'\D', '')
df_clean['technical_power'] = df_clean['technical_power'].str.replace(r'\D', '')
df_clean.head()
```

```python
df_marque=df_clean['url'].str.split('/1',expand=True)
df_clean['marque']=df_marque[0]
df_clean.drop(columns =["url"], inplace = True)

df_marque=df_clean['marque'].str.split('/',expand=True)
df_clean['marque']=df_marque[0]
df_clean['modele']=df_marque[1]

df_year=df_clean['year'].str.split(' le',expand=True)
df_clean['date_fabrication']=df_year[0]
df_clean.drop(columns =["year"], inplace = True)

df_clean['date_fabrication'] = df_clean['date_fabrication'].str.replace(r'\D', '')

df_clean.head()
```

```python
df_clean['actual_power']=df_clean['actual_power'].replace('N/A', 0)
df_clean=df_clean.replace('N/A', np.nan)
df_clean=df_clean.replace('', 0)
df_clean['emission']=df_clean['emission'].replace('nc', 0)
df_clean.head()
```

```python
df_clean=df_clean.rename(columns={"location": "code_postal", "price": "prix_vente", "mileage": "kilometrage",
 "fuel_type": "energie", "fuel_usage": "consommation", "door_nb": "portes",
  "technical_power": "cv_fiscaux", "actual_power": "cv_reels", "body_colour": "couleur", "body_type": "carrosserie", "seat_nb": "sieges"})
df_clean.head()
```

```python
df_clean.dtypes
```

```python
df_clean['emission']=df_clean['emission'].astype(float)
df_clean['consommation']=df_clean['consommation'].astype(float)
df_clean=df_clean.astype({'emission': 'int','kilometrage': 'int','prix_vente': 'int','code_postal': 'int','portes': 'int', 'cv_fiscaux': 'int', 'cv_reels': 'int', 'date_fabrication': 'int', 'sieges': 'int'})
df_clean.dtypes
```

```python
df_clean = df_clean[['version','marque','modele','date_fabrication','kilometrage','couleur','carrosserie'
,'code_postal','energie','emission','consommation','transmission','portes','sieges','cv_fiscaux','cv_reels','prix_vente']]
df_clean.head(10)
```
